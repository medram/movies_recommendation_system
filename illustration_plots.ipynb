```python
import numpy as np
import matplotlib.pyplot as plt

#plt.style.use("ggplot")
```

```python
x = [np.random.uniform() for _ in range(100)]
y = [np.random.uniform() for _ in range(100)]
```

```python
user_x = np.random.uniform(0.3, 0.7, 20)
user_y = np.random.uniform(0.3, 0.7, 20)

x_min = np.min(user_x)
y_min = np.min(user_y)

x_max = np.max(user_x)
y_max = np.max(user_y)
```

```python
from matplotlib.patches import Rectangle


plt.scatter(x, y, s=10, label="Movies")
plt.scatter(user_x, user_y, s=13, label="User's favorite movies")

# draw lines
# plt.axhline(y_max, xmin=x_min, xmax=x_max, color="orange", linestyle="--")
# plt.axhline(y_min, xmin=x_min, xmax=x_max, color="orange", linestyle="--")

# plt.axvline(x_max, ymin=y_min, ymax=y_max, color="orange", linestyle="--")
# plt.axvline(x_min, ymin=y_min, ymax=y_max, color="orange", linestyle="--")

# draw a rectangle
rect = Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, linewidth=1.5, edgecolor='orange', facecolor='none', linestyle="--")
plt.gca().add_patch(rect)

plt.title("User's movie interest space (just illustration)")
plt.grid(False)
plt.xlabel("(figure 1)")
plt.legend()
plt.savefig("imgs/figure_1.png")
```

```python
user_x = np.random.uniform(0.1, 0.4, 20)
user_y = np.random.uniform(0.1, 0.4, 20)

user_x = np.append(user_x, np.random.uniform(0.7, 0.9, 15))
user_y = np.append(user_y, np.random.uniform(0.7, 0.9, 15))


x_min = np.min(user_x)
y_min = np.min(user_y)

x_max = np.max(user_x)
y_max = np.max(user_y)
```

```python
from matplotlib.patches import Rectangle


plt.scatter(x, y, s=10, label="Movies")
plt.scatter(user_x, user_y, s=13, label="User's favorite movies")

# draw a rectangle
rect = Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, linewidth=1.5, edgecolor='orange', facecolor='none', linestyle="--")
plt.gca().add_patch(rect)

plt.title("User's movie interest space (just illustration)")
plt.grid(False)
plt.xlabel("(figure 2)")
plt.legend()
plt.savefig("imgs/figure_2.png")
```

```python
user_x1 = user_x[user_x < 0.5]
user_x2 = user_x[user_x >= 0.5]

user_y1 = user_y[user_y < 0.5]
user_y2 = user_y[user_y >= 0.5]

circle1_x = user_x1.mean()
circle1_y = user_y1.mean()

circle2_x = user_x2.mean()
circle2_y = user_y2.mean()
```

```python
from matplotlib.patches import Circle


plt.scatter(x, y, s=10, label="Movies")
plt.scatter(user_x, user_y, s=13, label="User's favorite movies")

# draw a rectangle
circle_1 = Circle((circle1_x, circle1_y), radius=0.2, edgecolor="orange", facecolor="none")
circle_2 = Circle((circle2_x, circle2_x), radius=0.14, edgecolor="orange", facecolor="none")

plt.gca().add_patch(circle_1)
plt.gca().add_patch(circle_2)

plt.title("User's movie interest spaces (just illustration)")
plt.grid(False)
plt.xlabel("(figure 3)")
plt.axis("square")
plt.legend()
plt.savefig("imgs/figure_3.png")
```
